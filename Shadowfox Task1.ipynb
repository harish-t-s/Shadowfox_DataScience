```python

import matplotlib.pyplot as plt

plt.plot([1,2,3,4,5],[3,1,4,5,2] )
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Shadowfox Line Chart')
plt.show()
print("This is a simple line chart using the matplotlib function")
```

```python
import matplotlib.pyplot as plt

x = [1, 2, 3, 4, 5]
y = [3, 1, 2, 7, 8]

plt.plot(x, y, marker='x')
plt.title('Frequency Detection')
plt.xlabel('Observation')
plt.ylabel('Occurance')
plt.show()
print("Line Graph is used for continuous values for example in weather forecasting")
```

```python
categories = ['A', 'B', 'C', 'D']
values = [90, 32, 78, 51]

plt.bar(categories, values)
plt.title('Student Academic Performance')
plt.xlabel('Student')
plt.ylabel('Marks')
plt.show()
print("Bar chart is used for comparing other attributes such as Student Academic performance")
```

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.random.normal(170, 10, 250)

plt.hist(x)
plt.show() 
print("histogram works well for continuous data and is associated with the weights used for comparision ")
print("Here np library function uses random function to obtain random values")
```

```python
sizes = [15, 30, 45, 10]
labels = ['A', 'B', 'C', 'D']

plt.pie(sizes, labels=labels, autopct='%1.1f%%')
plt.title('Pie Chart')
plt.show()
```
